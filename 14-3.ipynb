```python
import pandas as pd
import matplotlib.pyplot as pt
import numpy as np
```

```python
df=pd.read_csv(r"Main_covid1.csv")
```

```python
pd.to_datetime(df["Date"]).dt.strftime('%Y-%m-%d')
```

```python
Total_Case_deth_Rate_of_JanToApr=MthJan_Apr.groupby("Country")[["Cases","Deaths"]].agg([sum])
Total_Case_deth_Rate_of_JanToApr
```

```python
Total_Case_deth_Rate_of_MayToDec=MthMay_Dec.groupby("Country")[["Cases","Deaths"]].agg([sum])
Total_Case_deth_Rate_of_MayToDec
```

2.What is the relation between vaccination rates and Number of new Cases

```python
df["Vaccination_rate"]=(df["Vaccinated"]/df["Population"])*100
```

```python
pt.scatter(df["Vaccination_rate"],df["Cases"])
pt.xlabel("Vaccination_rate")
pt.ylabel("New_Cases")
pt.title("Relation_between_VaccinationRate And New Cases") 
pt.show()
```

3.Are there any trends in recovery rates across different counteries?

```python
df["Recovery_Rate"]=(df["Recovered"]/df["Active"])*100
```

```python
df.replace([np.inf, -np.inf], 1, inplace=True)
df["Recovery_Rate"]=df["Recovery_Rate"].astype(int)
```

```python
diffn_countery=df.groupby("Country")[["Recovery_Rate"]].sum()
diffn_countery                     
```

```python
pt.bar(df["Country"],df["Recovery_Rate"])
pt.show()
```
